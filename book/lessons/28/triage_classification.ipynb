
### EX: Text Classification

In this exercise, we will convert text into mathematical arrays (word counts) and train a Naive Bayes model. You will execute the following steps:

* **Generate Data:** Create a synthetic dataset of tactical text messages labeled as Routine or Urgent.
* **Vectorize Text:** Use a `CountVectorizer` to turn sentences into tables of word frequencies (the "Bag of Words" approach).
* **Train Model:** Fit a `MultinomialNB` (Multinomial Naive Bayes) model to the word counts.
* **Predict and Visualize:** Predict the class of new interceptions and print the exact mathematical probabilities the model calculated.

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# 1. Generate Synthetic Tactical Communications
data = {
    'Transmission': [
        "supply truck arriving at 0800", "weather clear for takeoff", 
        "enemy troops spotted at grid alpha", "routine perimeter check complete",
        "taking heavy fire request immediate support", "chow hall opens at 1700",
        "contact contact contact multiple hostiles", "refueling complete"
    ],
    'Label': ['Routine', 'Routine', 'Urgent', 'Routine', 'Urgent', 'Routine', 'Urgent', 'Routine']
}
df_comms = pd.DataFrame(data)

print("Intercepted Transmissions Dataset:")
print(df_comms, "\n")

# 2. Vectorize the text (Convert words to numbers)
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(df_comms['Transmission'])

# 3. Train the Naive Bayes Model
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train, df_comms['Label'])

# 4. Test on a New Interception
new_interception = ["spotted enemy troops taking fire"]
X_test = vectorizer.transform(new_interception)

# Predict class and probabilities
prediction = nb_classifier.predict(X_test)[0]
probabilities = nb_classifier.predict_proba(X_test)[0]

print(f"New Interception: '{new_interception[0]}'")
print(f"Predicted Classification: {prediction.upper()}")
print(f"Confidence (Probability of Routine): {probabilities[0]:.4f}")
print(f"Confidence (Probability of Urgent): {probabilities[1]:.4f}")

```


### Interpreting the Results

The text vectorizer strips the grammar away entirely, treating the sentence purely as a "bag of words." When the model sees the new string "spotted enemy troops taking fire", it looks up the historical likelihood of each individual word. Because "enemy", "troops", and "fire" appeared exclusively in "Urgent" training examples, the math overwhelmingly tilts the posterior probability toward Urgent, demonstrating how fast and effective this algorithm is for text triage.
