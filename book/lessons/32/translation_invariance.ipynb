### EX: Spatial Hierarchy and Dimensionality

In this exercise, we will build the foundation of a modern CNN architecture using Keras to observe how convolutions and pooling reduce data dimensions to achieve translation invariance.

* **Build Architecture:** Create a sequential model with a `Conv2D` layer followed by a `MaxPooling2D` layer.
* **Define Input:** We will specify an input shape of 64x64 pixels with 3 color channels (RGB).
* **Trace the Shape:** Use `model.summary()` to mathematically trace how the spatial dimensions shrink while the feature depth increases.



```python
#Only run this cell after selecting your kernel
!python.exe -m pip install --upgrade pip
!pip install numpy matplotlib scikit-learn pandas tensorflow
```

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# 1. Build a simple CNN Architecture
model = Sequential([
    # Input: 64x64 image with 3 color channels (RGB)
    # Filter: We ask the network to learn 16 different 3x3 feature filters
    Conv2D(filters=16, kernel_size=(3, 3), padding='same', activation='relu', input_shape=(64, 64, 3)),
    
    # Pooling: A 2x2 window that slides with a stride of 2
    MaxPooling2D(pool_size=(2, 2), strides=2),
    
    # Add a deeper layer to find more complex features
    Conv2D(filters=32, kernel_size=(3, 3), padding='same', activation='relu'),
    MaxPooling2D(pool_size=(2, 2), strides=2),
    
    # Flatten the 2D spatial maps into a 1D vector for final classification
    Flatten(),
    Dense(1, activation='sigmoid')
])

# 2. Inspect the Spatial Transformation
print(model.summary())
```


### Interpreting the Results

Run the code and look at the `Output Shape` column in the console.

1. The first `Conv2D` outputs `(None, 64, 64, 16)`. Because we used `padding='same'`, the 64x64 spatial dimensions were preserved, but the image was split into 16 distinct "Feature Maps" (e.g., one map for vertical edges, one for red blobs, etc.).

2. The `MaxPooling2D` layer outputs `(None, 32, 32, 16)`. The 2x2 pooling window perfectly chopped the spatial dimensions in half.

3. By the time the data hits the `Flatten` layer, the 64x64 image has been compressed into tiny 16x16 grids, but we have 32 deeply profound feature maps. The network has successfully exchanged raw spatial resolution for deep tactical understanding.
